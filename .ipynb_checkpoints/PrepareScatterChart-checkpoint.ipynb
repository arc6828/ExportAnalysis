```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("http://export.ckartisan.com/api/export-2hs?year_at=2019")
df
```

```python
df_pivot = pd.pivot_table(df, values='value', index=['exporter'], columns=['hscode'], aggfunc=np.sum)
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
df_pivot.isnull().sum().sum()
```

```python
X = df_pivot.to_numpy()
X
```

```python
model = TSNE(n_components=2, random_state=0)
U = model.fit_transform(X) 
U
```

```python
list(U[:,0])
```

```python
type(U)
```

```python
countries = list(df_pivot.axes[0])
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(list(U[:,0]), list(U[:,1]), color='r', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('export every country in 2019')

plt.text(0.55, 0.6, "spam", size=10, rotation=0.,
         ha="right", va="top",
         bbox=None
         )


plt.show()
```
