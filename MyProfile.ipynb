```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
X2 = pd.read_csv('data/export-hs4-pivot.csv')
#X_new => 2019
X2_new = X2.loc[X2['year_at'] == 2019]
X2_new = X2_new.drop(['year_at'], axis=1)
X2_new = X2_new.set_index(['exporter'])
```

```python
Thailand = json.loads(X2_new.loc['Thailand'].to_json())
list(Thailand.keys())
hs2_list = np.array(list(map(lambda x: x[0]+x[1], list(Thailand.keys()))))
ft = (hs2_list == "71")

fig, axs = plt.subplots(4, 1, figsize=(30, 15), sharey=True)
#THAILAND
Thailand = json.loads(X2_new.loc['Thailand'].to_json())
axs[0].set_title('Thailand')
axs[0].bar(np.array(list(Thailand.keys()))[ft], np.array(list(Thailand.values()))[ft] )


#Switzerland
Switzerland = json.loads(X2_new.loc['Switzerland'].to_json())
axs[1].set_title('Switzerland')
axs[1].bar(np.array(list(Switzerland.keys()))[ft], np.array(list(Switzerland.values()))[ft] )


#Malaysia
Malaysia = json.loads(X2_new.loc['Malaysia'].to_json())
axs[2].set_title('Malaysia')
axs[2].bar(np.array(list(Malaysia.keys()))[ft], np.array(list(Malaysia.values()))[ft] )


#Singapore
Singapore = json.loads(X2_new.loc['Singapore'].to_json())
axs[3].set_title('Singapore')
axs[3].bar(np.array(list(Singapore.keys()))[ft], np.array( list(Singapore.values()))[ft] )

fig.suptitle('Cluster Analysis : Thailand, Switzerland, Malaysia, and Singapore ')
```
